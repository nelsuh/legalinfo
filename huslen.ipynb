```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage you want to extract text from
textsmon = []
for link in links:
    
    url = link  # Get the URL at index i

    # Make a GET request to fetch the page content of the current URL
    response = requests.get(url)
    try:
        response.raise_for_status()  # Check if the request was successful
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract all text from the webpage
        divs = soup.find_all('div', class_='maincontenter w-100 pull-left')
        text = ' '.join(div.get_text(separator=' ', strip=True) for div in divs)
        text = text.replace("Хэвлэх", "")
        start_index = text.find('1 дүгээр зүйл.')
        if start_index != 1:
            text =text[start_index:]
            textsmon.append(text)
        # Print the extracted text
        print(f"Text from {url}:")
        print(text)
        print("\n---\n")  # Separator between outputs for different URLs
        
    except requests.exceptions.HTTPError as e:
        # If the request was not successful, print the error
        print(f"Failed to retrieve data from {url}. Error: {e}")
```

```python
import json

def write_list_to_jsonl(all1, output_file_path):
    # Open the output file in write mode
    with open(output_file_path, 'w', encoding='utf-8') as file:
        # Iterate over each item in the list
        for mon_text in all1:
            for text in mon_text:
                # Create a dictionary with the key "mon"
                data = {'mon': text}
                # Write the dictionary as a JSON-formatted string to the file
                file.write(json.dumps(data,ensure_ascii=False) + '\n')


# Specify the output file path
output_file_path = 'mongolian_texts.jsonl'

# Call the function to write the list to a JSONL file
write_list_to_jsonl(all1, output_file_path)

# This will create a file 'mongolian_texts.jsonl' in the specified directory
```
